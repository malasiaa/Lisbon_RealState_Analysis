```python
def remove_non_numeric_and_convert(value):
    if value.strip() == '':
        return np.nan  # or any other default value you prefer
    numeric_value = re.sub(r'[^0-9]', '', value)
    numeric_value = numeric_value.replace('.', '')
    numeric_value = numeric_value.replace(',', '')
    return float(numeric_value)


result['preço'] = result['preço'].apply(remove_non_numeric_and_convert)


print(result)
```
