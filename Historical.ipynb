```python


data = pd.read_excel('INEdata.xlsx', sheet_name='MvsTipology')


location = 'Lisboa'
dwelling_typologies = ['0 or 1 bedroom', '2 bedrooms', '3 bedrooms', '4 or more bedrooms']
filtered_data = data[(data['Location'] == location) & (data['Dwelling typology'].isin(dwelling_typologies))]

filtered_data
```

## Median value sales per m<sup>2</sup> dwelling typology 
    
</div>

```python
plt.figure(figsize=(10, 6))  


quarters = filtered_data.columns[2:]  
years = ['2019', '2020', '2021', '2022']  


sorted_quarters = []
for year in years:
    sorted_quarters.extend([f'{quarter} {year}' for quarter in ['1stQ', '2ndQ', '3rdQ', '4thQ'] if f'{quarter} {year}' in quarters])


for typology in dwelling_typologies:
    typology_data = filtered_data[filtered_data['Dwelling typology'] == typology]
    typology_values = [typology_data.iloc[0, typology_data.columns.get_loc(quarter)] for quarter in sorted_quarters]
    plt.plot(sorted_quarters, typology_values, marker='o', label=typology)


plt.xlabel('Quarter')
plt.ylabel('Median value sale per Dwellings m2')
plt.title(f'Median value sales m2 of Dwellings {location}')

plt.xticks(rotation=45)


plt.legend()

plt.show()
```

```python
effort=pd.merge(merged_df, gains,left_index=True, right_index=True, how='outer')
```

```python
effort
```

```python
effort['Lisboa Salary Growth'] = effort['Lisboa mean Salary'].pct_change() * 100
effort['Portugal Salary Growth'] = effort['Portugal mean Salary'].pct_change() * 100
effort['Lisboa Gains Growth'] = effort['Lisboa mean Gains'].pct_change() * 100
effort['Portugal Gains Growth'] = effort['Portugal mean Gains'].pct_change() * 100
effort['Lisboa House Growth'] = effort['Lisboa'].pct_change() * 100
effort['Portugal House Growth'] = effort['Portugal'].pct_change() * 100
effort['Lisboa Salary Cum Growth'] = effort['Lisboa mean Salary'].pct_change().cumsum() * 100
effort['Portugal Salary Cum Growth'] = effort['Portugal mean Salary'].pct_change().cumsum() * 100
effort['Lisboa Gains Cum Growth'] = effort['Lisboa mean Gains'].pct_change().cumsum() * 100
effort['Portugal Gains Cum Growth'] = effort['Portugal mean Gains'].pct_change().cumsum() * 100
effort['Lisboa House Cum Growth'] = effort['Lisboa'].pct_change().cumsum() * 100
effort['Portugal House Cum Growth'] = effort['Portugal'].pct_change().cumsum() * 100
effort
```

```python
effort = effort.reset_index()
effort = effort.rename(columns={'index': 'Anos'})
```

```python
effort
```

```python
effort
```

```python
import matplotlib.ticker as ticker
plt.figure(figsize=(10, 6))
plt.plot(effort['Anos'], effort['Lisboa House Cum Growth'], label='Lisboa Cum House Growth')
plt.plot(effort['Anos'], effort['Lisboa Salary Cum Growth'], label='Lisboa Cum Salary Growth')
plt.plot(effort['Anos'], effort['Lisboa Gains Cum Growth'], label='Lisboa Cum Gains Growth')

plt.xlabel('Year')
plt.ylabel('Percentage Growth')
plt.title('Lisboa House Growth vs Lisboa Salary Growth and Lisboa Gains Growth')
plt.legend()
plt.grid(True)
plt.xticks(effort['Anos'])
ax = plt.gca()
ax.yaxis.set_major_locator(ticker.MultipleLocator(base=10))
ax.yaxis.set_major_formatter(ticker.PercentFormatter())
plt.show()







```

## Number of Lisbon House Contracts
    
</div>


```python
contracts=pd.read_excel(r'C:\Users\Utilizador\Downloads\contracts.xlsx',usecols="A,C,E,G,I,K,M,O,Q,S,U,W", skiprows=10, nrows=3)
contracts
```

```python
for i, column in enumerate(contracts.columns):
    if column.startswith('N.º'):
        new_column_name = str(2020 - i)
        contracts.rename(columns={column: new_column_name}, inplace=True)

contracts = contracts.rename(columns={'Unnamed: 0': 'Location'})        
contracts
```

```python
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, find_peaks

lisboa_data = contracts[contracts['Location'] == 'Lisboa']

years = contracts.columns[1:][::-1]
values = lisboa_data.iloc[:, 1:].values.flatten()[::-1]

# Find the inflection points
peaks, _ = find_peaks(values)
lows, _ = find_peaks(-values)

plt.plot(years, values, marker='o',color = "black")
plt.plot(years[peaks], values[peaks], 'ro')  
plt.plot(years[lows], values[lows], 'ro') 
plt.title('Number of Housing Contracts (2009-2019)')
plt.xlabel('Year')
plt.ylabel('Contracts')
plt.show()








```
